```python
loc = input('Enter image location :')
img = cv2.imread(loc)
if img is not None:
    c = int(input('Enter 0 for show and 1 for save the image'))
    if c==0:
        c2 = int(input('Enter 0 for Black&White and 1 for as it is'))
        if c2 == 0:
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            cv2.imshow('Grayscale image',gray)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        else:
            cv2.imshow('Image showing',img)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
    else:
        cv2.imwrite('output_img.jpg',img)
        print('Image saved successfull as output_img')
else:
    print('Could not load image')
```
