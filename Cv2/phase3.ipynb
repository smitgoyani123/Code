#  Phase 3: Basic Image Drawing Techniques

```python
import cv2
```

```python
image = cv2.imread('python_img.png')

if image is None:
    print('Image is not found')
else:
    print('Image loaded successfully')

    pt1 = (50,90)
    pt2 = (70,20)
    color = (255,0,0)
    thickness = 2
    cv2.line(image,pt1,pt2,color,thickness)
    cv2.imshow('Line Drawing',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if image is None:
    print('Image is not found')
else:
    print('Image loaded successfully')
    pt1 = (50,50)
    pt2 = (100,120)
    color = (0,255,0)
    thickness = 3
    cv2.rectangle(image,pt1,pt2,color,thickness)
    cv2.imshow('Rectangle Drawing',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if image is None:
    print('Image is not found')
else:
    print('Image loaded successfully')
    cv2.circle(image,(50,50),50,(0,0,255),5)
    cv2.imshow('Circle Drawing',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if image is None:
    print('Image is not found')
else:
    print('Image loaded successfully')
    cv2.putText(image,'Hello Python Programmer',(10,50),cv2.FONT_HERSHEY_COMPLEX,1.2,(0,0,255),2)
    cv2.imshow('Adding text over image',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
