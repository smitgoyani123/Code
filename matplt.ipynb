# Pyplot

```python
import matplotlib.pyplot as plt

x = ['Mon','Tue','Wed','Thu','Fri']
y = [10,15,7,20,12]

plt.plot(x,y)
plt.title("Backery sales this week!")
plt.xlabel('Day of this week')
plt.ylabel('Sales per day')
plt.show()
```

```python
#plt.plot(x, y, color='color name',lineStyle='line_style',linewidth=value,marker='marker symbole',label='label name')
months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months,sales,color='blue',linestyle='--',linewidth=2,marker='o',label='2025 sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Data Report')
plt.legend(loc='upper left',fontsize=12)  #upper-left corner title

#Grid 
plt.grid(color='gray',linestyle=':',linewidth=1)

#Axis Limit 
plt.xlim(1,4)
plt.ylim(0,2000)

#Axis ticks
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

 ## Bargraph

```python
#plt.bar(x,height, color='color name',width= value,label= 'label name')

product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product,sales,color='purple',label='Sales 2025') #horizontal
#vertical - barh

plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparision')
plt.legend()
plt.show()
```

## Piechart

```python
#plt.pie(values,labels=label_list,colors=color_list,autopct='%1.1f%%')

regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,colors=['gold','skyblue','lightgreen','coral'],autopct='%1.1f%%')
plt.title('Revenue Contribution by Regions')
plt.show()
```

## Histograms

```python
#plt.hist(data,bins=number_of_bins,color='colorname',edgecolor= 'black')
scores = [45,87,34,78,40,23,56,73,19,64,28,11,78,88,93,12,13,18,55,59]

plt.hist(scores,bins=5,color='orange',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score Distributions of students')
plt.show()
#hight of bar = no of students in that range
```

## Scatter

```python
#plt.scatter(x,y,color='color_name',marker='marker_style',label='label name')

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies,exam_scores,color='red',marker='^',label='Student Data')

plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title('Relation between study time and Exam scores')
plt.legend()
plt.grid(True)
plt.show()
```

## Subplot

```python
#plt.subplot(nrows,ncols,index)
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar chart')

plt.tight_layout()
plt.show()

```

```python
#fig, ax = plt.subplot(nrows,ncols,figsize=(width,height))
fig, ax = plt.subplots(1,2,figsize=(10,5))
x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title('Line plot')

ax[1].bar(x,y)
ax[1].set_title('Bar chart')

fig.suptitle('Comparison of Line and Bar chart')
plt.tight_layout()

fig, ax = plt.subplots(2,1,sharex = True)
plt.show()


```
